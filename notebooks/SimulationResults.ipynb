In this notebook, we show the results of the evolution simulation.

```python
from asymmetree.visualize.TreeVis import visualize, assign_colors, assign_gene_colors, assign_species_colors
from tralda.datastructures import Tree
import matplotlib.pyplot as plt

species_tree = Tree.load("../data/raw/species_tree.json")
```

```python
dict_colors = assign_species_colors(species_tree)
visualize(species_tree,color_dict=dict_colors, save_as="../reports/figures/species_tree.png")
```

```python
import os

genes_dir = "../data/raw/true_gene_trees/"
for i, file in enumerate(os.listdir(genes_dir)):
    if i >= 50:
        break
    gene_tree = Tree.load(genes_dir + file)
    
    visualize(gene_tree)
```
